л Ҳмқҙлё” мқёмҪ”л”©

```python
from sklearn.preprocessing import LabelEncoder

items = ['TV', 'лғүмһҘкі ', 'м „мһҗл Ҳмқём§Җ', 'м»ҙн“Ён„°', 'м„ н’Қкё°', 'м„ н’Қкё°', 'лҜ№м„ң', 'лҜ№м„ң']

# LabelEncoderлҘј к°қмІҙлЎң мғқм„ұн•ң нӣ„, fit()кіј transform()мңјлЎң л Ҳмқҙлё” мқёмҪ”л”© мҲҳн–ү
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
print('мқёмҪ”л”© ліҖнҷҳк°’:', labels)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

items = ['TV', 'лғүмһҘкі ', 'м „мһҗл Ҳмқём§Җ', 'м»ҙн“Ён„°', 'м„ н’Қкё°', 'м„ н’Қкё°', 'лҜ№м„ң', 'лҜ№м„ң']

# 2м°Ёмӣҗ ndarrayлЎң ліҖнҷҳн•©лӢҲлӢӨ.
items = np.array(items).reshape(-1,1)

# мӣҗ-н•« мқёмҪ”л”©мқ„ м Ғмҡ©н•©лӢҲлӢӨ
oh_encoder = OneHotEncoder()
oh_encoder.fit(items)
oh_labels = oh_encoder.transform(items)

# OneHotEncoderлЎң ліҖнҷҳн•ң кІ°кіјлҠ” нқ¬мҶҢн–үл ¬мқҙлҜҖлЎң toarray()лҘј мқҙмҡ©н•ҙ л°Җм§‘ н–үл ¬лЎң ліҖнҷҳ
print('мӣҗ-н•« мқёмҪ”л”© лҚ°мқҙн„°')
print(oh_labels.toarray())
print('мӣҗ-н•« мқёмҪ”л”© лҚ°мқҙн„° м°Ёмӣҗ')
print(oh_labels.shape)
```

```python
from sklearn.datasets import load_iris
import pandas as pd
# л¶“кҪғ лҚ°мқҙн„° м„ёнҠёлҘј лЎңл”©н•ҳкі  DataFrameмңјлЎң ліҖнҷҳн•©лӢҲлӢӨ
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data = iris_data, columns = iris.feature_names)

print('feature л“Өмқҳ нҸүк·  к°’')
print(iris_df.mean())
print('\nfeature л“Өмқҳ л¶„мӮ° к°’')
print(iris_df.var())
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler к°қмІҙ мғқм„ұ
scaler = StandardScaler()
# StandardScalerлЎң лҚ°мқҙн„° м„ёнҠё ліҖнҷҳ. fit()кіј transform() нҳём¶ң
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

# transform() мӢң мҠӨмјҖмқј ліҖнҷҳлҗң лҚ°мқҙн„° м„ёнҠёк°Җ NumPy ndarrayлЎң л°ҳнҷҳлҸј мқҙлҘј DataFrameмңјлЎң л°ҳнҷҳ
iris_df_scaled =pd.DataFrame(data = iris_scaled, columns = iris.feature_names)
print('feature л“Өмқҳ нҸүк·  к°’')
print(iris_df_scaled.mean())
print('\nfeature л“Өмқҳ л¶„мӮ° к°’')
print(iris_df_scaled.var())
```

MinMaxScaler:

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler к°қмІҙ мғқм„ұ
scaler = MinMaxScaler()
# MinMaxScalerлЎң лҚ°мқҙн„° м„ёнҠё ліҖнҷҳ. fit()кіј transform() нҳём¶ң.
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

# transform()мӢң мҠӨмјҖмқј ліҖнҷҳлҗң лҚ°мқҙн„° м„ёнҠёк°Җ Numpy ndarrayлЎң л°ҳнҷҳлҸј мқҙлҘј DataFrameмңјлЎң ліҖнҷҳ
iris_df_scaled = pd.DataFrame(data = iris_scaled, columns = iris.feature_names)
print('featureл“Өмқҳ мөңмҶҹк°’')
print(iris_df_scaled.min())
print('\nfeatureл“Өмқҳ мөңлҢ“к°’')
print(iris_df_scaled.max())
```

н•ҷмҠө лҚ°мқҙн„°мҷҖ н…ҢмҠӨнҠё лҚ°мқҙн„°мқҳ мҠӨмјҖмқјл§Ғ ліҖнҷҳ мӢң мЈјмқҳм җ
1. к°ҖлҠҘн•ҳлӢӨл©ҙ м „мІҙ лҚ°мқҙн„°мқҳ мҠӨмјҖмқјл§Ғ ліҖнҷҳмқ„ м Ғмҡ©н•ң л’Ө н•ҷмҠөкіј нғңмҠӨнҠё лҚ°мқҙн„°лЎң л¶„лҰ¬
2. 1мқҙ м—¬мқҳм№ҳ м•ҠлӢӨл©ҙ н…ҢмҠӨнҠё лҚ°мқҙн„° ліҖнҷҳ мӢңм—җлҠ” fit()мқҙлӮҳ fit_transform()мқ„ м Ғмҡ©н•ҳм§Җ м•Ҡкі  н•ҷмҠө лҚ°мқҙн„°лЎң мқҙлҜё fit()лҗң Scaler к°қмІҙлҘј мқҙмҡ©н•ҙ transform()мңјлЎң ліҖнҷҳ

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# н•ҷмҠө лҚ°мқҙн„°лҠ” 0л¶Җн„° 10к№Ңм§Җ, н…ҢмҠӨнҠё лҚ°мқҙн„°лҠ” 0л¶Җн„° 5к№Ңм§Җ к°’мқ„ к°Җм§ҖлҠ” лҚ°мқҙн„° м„ёнҠёлЎң мғқм„ұ
# Scaler нҒҙлһҳмҠӨмқҳ fit(), transform()мқҖ 2м°Ёмӣҗ мқҙмғҒ лҚ°мқҙн„°л§Ң к°ҖлҠҘн•ҳлҜҖлЎң reshape(-1,1)лЎң м°ЁмӣҗліҖкІҪ
train_array = np.arange(0, 11).reshape(-1,1)
test_array = np.arange(0,6).reshape(-1,1)
```

```python
# MinMaxScaler к°қмІҙм—җ лі„лҸ„мқҳ feature_range нҢҢлқјлҜён„° к°’мқ„ м§Җм •н•ҳм§Җ м•Ҡмңјл©ҙ 0~1 к°’мңјлЎң ліҖнҷҳ
scaler = MinMaxScaler()

# fit()н•ҳкІҢ лҗҳл©ҙ train_array лҚ°мқҙн„°мқҳ мөңмҶҹк°’мқҙ 0, мөңлҢҖк°’мқҙ 10мңјлЎң м„Өм •
scaler.fit(train_array)

# 1/10 scaleлЎң train_array лҚ°мқҙн„° ліҖнҷҳн•Ё. мӣҗліё 10->1лЎң ліҖнҷҳлҗЁ
train_scaled = scaler.transform(train_array)

print('мӣҗліё train_array лҚ°мқҙн„°:', np.round(train_array.reshape(-1),2))
print('Scaleлҗң train_array лҚ°мқҙн„°:', np.round(train_array.reshape(-1),2))
```

```python
# MinMaxScalerм—җ test_arrayлҘј fit()н•ҳкІҢ лҗҳл©ҙ мӣҗліё лҚ°мқҙн„°мқҳ мөңмҶҹк°’мқҙ 0, мөңлҢ“к°’мқҙ 5лЎң м„Өм •лҗЁ
scaler.fit(test_array)

# 1/5 scaleлЎң train_array лҚ°мқҙн„° ліҖнҷҳн•Ё. мӣҗліё 5->1лЎң ліҖнҷҳлҗЁ
test_scaled = scaler.transform(test_array)

# test_arrayмқҳ scale ліҖнҷҳ м¶ңл Ҙ
print('мӣҗліё train_array лҚ°мқҙн„°:', np.round(train_array.reshape(-1),2))
print('Scaleлҗң train_array лҚ°мқҙн„°:', np.round(train_array.reshape(-1),2))
```

```python
scaler = MinMaxScaler()
scaler.fit(train_array)
train_scaled = scaler.transform(train_array)
print('мӣҗліё train_array лҚ°мқҙн„° :', np.round(train_array.reshape(-1),2))
print('Scaleлҗң train_array лҚ°мқҙн„°:', np.round(train_scaled.reshape(-1),2))

# test_arrayм—җ Scale ліҖнҷҳмқ„ н•  л•ҢлҠ” л°ҳл“ңмӢң fit()мқ„ нҳём¶ңн•ҳм§Җ м•Ҡкі  transform()л§ҢмңјлЎң ліҖнҷҳн•ҙм•јн•Ё
test_scaled = scaler.transform(test_array)
print('\nмӣҗліё test_array лҚ°мқҙн„°:', np.round(test_array.reshape(-1),2))
print('Scaleлҗң test_array лҚ°мқҙн„°:', np.round(test_scaled.reshape(-1),2))
```

```python
# мһ…л Ҙ ageм—җ л”°лқј кө¬л¶„ к°’мқ„ л°ҳнҷҳн•ҳлҠ” н•ЁмҲҳ м„Өм •. DataFrameмқҳ apply lambda мӢқм—җ мӮ¬мҡ©
def get_category(age):
  cat = ''
  if age <= -1: cat= 'Unknown'
  elif age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young adult'
  elif age <= 60: cat = 'Adult'
  else: cat = 'Elderly'

  return cat

# л§үлҢҖк·ёлһҳн”„мқҳ нҒ¬кё° figureмқ„ лҚ” нҒ¬кІҢ м„Өм •
plt.figure(figsize = (10,6))

# Xм¶•мқҳ к°’мқ„ мҲңм°Ём ҒмңјлЎң н‘ңмӢңн•ҳкё° мң„н•ң м„Өм •
group_means = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young adult', 'Adult', 'Elderly']

# lambda мӢқм—җ мң„м—җм„ң мғқм„ұн•ң get_category() н•ЁмҲҳлҘј л°ҳнҷҳ к°’мңјлЎң м§Җм •
# get_category(x)лҠ” мһ…л Ҙк°’мңјлЎң 'Age' м№јлҹј к°’мқ„ л°ӣм•„м„ң н•ҙлӢ№н•ҳлҠ” cat л°ҳнҷҳ
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
sns.barplot(x='Age_cat', y='Survived', hue='Sex', data=titanic_df, order=group_means)
titanic_df.drop('Age_cat', axis=1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

def encoder_features(dataDF):
  features = ['Cabin', 'Sex', 'Embarked']
  for feature in features:
    le = LabelEncoder()
    le = le.fit(dataDF[feature])
    dataDF[feature] = le.transform(dataDF[feature])

  return dataDF

titanic_df = encoder_features(titanic_df)
titanic_df.head()
```

```python
# NULL мІҳлҰ¬ н•ЁмҲҳ
def fillna(df):
  df['Age'].fillna(df['Age'].mean(), inplace = True)
  df['Cabin'].fillna('N', inplace = True)
  df['Embarked'].fillna('N', inplace = True)
  df['Fare'].fillna(0, inplace = True)

  return df

# лЁёмӢ лҹ¬лӢқ м•Ңкі лҰ¬мҰҳм—җ л¶Ҳн•„мҡ”н•ң н”јмІҳ м ңкұ°
def drop_features(df):
  df.drop(['PassengerId', 'Name', 'Ticket'], axis = 1, inplace= True)
  return df

# л Ҳмқҙлё” мқёмҪ”л”© мҲҳн–ү
def format_features(df):
  df['Cabin'] = df['Cabin'].str[:1]
  features = ['Cabin', 'Sex', 'Embarked']
  for feature in features:
    le = LabelEncoder()
    le = le.fit(df[feature])
    df[feature] = le.transform(df[feature])
  return df

  # м•һм—җм„ң м„Өм •н•ң лҚ°мқҙн„° м „мІҳлҰ¬ н•ЁмҲҳ нҳём¶ң
  def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)

    return df
```

```python
# мӣҗліё лҚ°мқҙн„°лҘј мһ¬лЎңл”©н•ҳкі , н”јмІҳ лҚ°мқҙн„° м„ёнҠёмҷҖ л Ҳмқҙлё” лҚ°мқҙн„° м„ёнҠё м¶”м¶ң
titanic_df = pd.read_csv('/content/drive/MyDrive/datasets/titanic_train.csv')
y_titanic_df = titanic_df['Survived']
X_titanic_df = titanic_df.drop('Survived', axis = 1)

X_titanic_df = transform_features(X_titanic_df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df, test_size = 0.2, random_state = 11)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# кІ°м •нҠёлҰ¬, Random Forest, лЎңм§ҖмҠӨнӢұ нҡҢк·ҖлҘј мң„н•ң мӮ¬мқҙнӮ·лҹ° Classifier нҒҙлһҳмҠӨ мғқм„ұ
dt_clf = DecisionTreeClassifier(random_state = 11)
rf_clf = RandomForestClassifier(random_state = 11)
lr_clf = LogisticRegression(solver = 'liblinear')

# DecisionTreeClassifier н•ҷмҠө/мҳҲмёЎ/нҸүк°Җ
dt_clf.fit(X_train, y_train)
dt_pred = dt_clf.predict(X_test)
print('DecisionTreeClassifier м •нҷ•лҸ„: {0:.4f}'.format(accuracy_score(y_test, dt_pred)))

# RandomForestClassifier н•ҷмҠө/м–ҳм¶•/нҸүк°Җ
rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
print('RandomForestClassifier м •нҷ•лҸ„: {0:.4f}'.format(accuracy_score(y_test, rf_pred)))

# LogisticRegression н•ҷмҠө/м–ҳм¶•/нҸүк°Җ
lr_clf.fit(X_train, y_train)
lr_pred = rf_clf.predict(X_test)
print('LogisticRegression м •нҷ•лҸ„: {0:.4f}'.format(accuracy_score(y_test, rf_pred)))
```

```python
from sklearn.model_selection import KFold

def exec_kfold(clf, folds = 5):
  # нҸҙл“ң м„ёнҠёлҘј 5к°ңмқё KFold к°қмІҙлҘј мғқм„ұ, нҸҙл“ң мҲҳл§ҢнҒј мҳҲмёЎ кІ°кіј м ҖмһҘмқ„ мң„н•ң лҰ¬мҠӨнҠё к°қмІҙ мғқм„ұ
  kfold = KFold(n_splits=folds)
  scores = []

  # KFold көҗм°Ё кІҖмҰқ мҲҳн–ү
  for iter_count, (train_index, test_index) in enumerate(kfold.split(X_titanic_df)):
    # X_titanic_df лҚ°мқҙн„°м—җм„ң көҗм°Ё кІҖмҰқлі„лЎң н•ҷмҠөкіј кІҖмҰқ лҚ°мқҙн„°лҘј к°ҖлҘҙнӮӨлҠ” index мғқм„ң
    X_train, X_test, = X_titanic_df.values[train_index], X_titanic_df.values[test_index]
    y_train, y_test = y_titanic_df.values[train_index], y_titanic_df.values[test_index]
    # Classifier н•ҷмҠө, мҳҲмёЎ м •нҷ•лҸ„ кі„мӮ°
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    scores.append(accuracy)
    print("көҗм°Ё кІҖмҰқ {0} м •нҷ•лҸ„: {1:.4f}".format(iter_count, accuracy))

    # 5к°ң foldм—җм„ңмқҳ нҸүк·  м •нҷ•лҸ„ кі„мӮ°
    mean_score = np.mean(scores)
    print("нҸүк·  м •нҷ•лҸ„: {0:.4f}".format(mean_score))

# exec_kfold нҳём¶ң
exec_kfold(dt_clf, folds = 5)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(dt_clf, X_titanic_df, y_titanic_df, cv = 5)

for iter_count, accuracy in enumerate(scores):
  print("көҗм°Ё кІҖмҰқ {0} м •нҷ•лҸ„ : {1:.4f}".format(iter_count, accuracy))

print("нҸүк·  м •нҷ•лҸ„ : {0:.4f}".format(np.mean(scores)))
```
